# Visualizing Rolling Statistics with Pandas DataFrames


## 1. Import Libraries and Dependencies

```python
# Import libraries and dependencies
import pandas as pd 
import pathlib as Path 
```

## 2. Create a Path to the File Using Pathlib.

```python
# Read the MSFT data, set the `date` as the index
MSFT_csv_path = (r"C:\Users\minha\OneDrive\Documents\ASU_2022_Bootcamp\Activity_Files\04-Pandas\2\Activities\09-rolling-statistics-01\Resources\MSFT.csv")
```

## 3. Read the CSV into a Pandas DataFrame.

```python
# Read the MSFT CSV into a DataFrame while formatting and setting the `date` column as the index
MSFT_df = pd.read_csv(MSFT_csv_path)
MSFT_df.set_index(pd.to_datetime(MSFT_df["date"], infer_datetime_format = True), inplace = True)
MSFT_df.head()
```

## 4. Drop Extra Columns

```python
# Use the `drop` function to drop the extra columns
MSFT_df.drop(["date", "volume", "open", "high", "low"], axis = "columns", inplace = True)
MSFT_df.head()
```

## 5. Plot Daily Closing Price of MSFT

```python
# Plot the daily closing prices of MSFT for the past 10 years
MSFT_df.plot(title = "Daily Closing Prices of MSFT", figsize = (15,5))
```

## 6. Rolling 7-Day Mean of MSFT Closing Prices

```python
# Plot a rolling 7-day mean of MSFT's closing prices
MSFT_df.rolling(window = 7).mean().plot(title = "7-day Mean of MSFT Closing Prices", figsize = (15, 5))
```

## 7. Rolling 7-Day Standard Deviation of MSFT Closing Prices

```python
# Plot a rolling 7-day std dev of MSFT's closing prices
MSFT_df.rolling(window = 7).std().plot(title = "7-day STD of MSFT Closing Prices", figsize = (15, 5))
```

## 8. Rolling 30-Day Mean and Standard Deviation of MSFT Closing Prices

```python
# Plot a rolling 30-day mean of MSFT's closing prices
MSFT_df.rolling(window = 30).mean().plot(title = "30-day Mean & STD of MSFT Closing Prices", figsize = (15,5))
```

## 9. Rolling 30-Day Standard Deviation of MSFT Closing Prices

```python
# Plot a rolling 30-day std dev of MSFT's closing prices
MSFT_df.rolling(window = 30).std().plot(title = "30-day STD of MSFT Closing Prices", figsize = (15,5))
```

## 10. Rolling 180-Day Mean of TSLA Closing Prices

```python
# Plot a rolling 180-day mean of MSFT's closing prices
MSFT_df.rolling(window = 180).mean().plot(title = "180-day Mean of MSFT Closing Prices", figsize = (15,5))
```

## 11. Rolling 180-Day Standard Deviation of TSLA Closing Prices

```python
# Plot a rolling 180-day std dev of MSFTs closing prices
MSFT_df.rolling(window = 180).std().plot(title = "180-day STD of MSFT Closing Prices", figsize = (15,5))
```

## 12. Overlay Daily Closing Price and 180-Day Rolling Mean of MSFT

```python
# Set figure of the daily closing prices of MSFT
ax = MSFT_df.plot() 

# Plot 180-Day Rolling Mean on the same figure
MSFT_df.rolling(window = 180).mean().plot(ax = ax, figsize = (15,5))

# Plot 30-Day Rolling Mean on the same figure
MSFT_df.rolling(window = 30).mean().plot(ax = ax, figsize = (15,5))

# Set the legend of the figure
ax.legend(["MSFT", "MSFT 180 Day Mean", "MSFT 30 Day Mean"])
```
