# Calculating Stock Returns using Pandas DataFrames

## 1. Import Libaries and Dependencies

```python
# Import necessary libraries and dependencies
import pandas as pd 
import pathlib as Path 
```

## 2. Create a Path to the File Using Pathlib

```python
# Use the Pathlib libary to set the path to the CSV
csv_path = (r"C:\Users\minha\OneDrive\Documents\ASU_2022_Bootcamp\Activity_Files\04-Pandas\2\Activities\08-returns-01\Resources\MSFT.csv")
```

## 3. Read CSV in as DataFrame

```python
# Read in the CSV as a DataFrame
MSFT_df = pd.read_csv(csv_path)
MSFT_df.head()
```

## 4. Drop Extra Columns

```python
# Drop unnecessary columns
MSFT_df = MSFT_df.drop(columns = ["volume", "open", "high", "low"])
MSFT_df.head()
```

## 5. Set Date as Index

```python
# Set the date as the index to the DataFrame
MSFT_df.set_index(pd.to_datetime(MSFT_df["date"], infer_datetime_format = True), inplace = True)
MSFT_df.head() 
```

## 6. Drop Extra Date Column

```python
# Drop the extra date column
MSFT_df = MSFT_df.drop(["date"], axis = 1)
MSFT_df.head()
```

## 7. Plot Daily Close

```python
# Plot the daily closing stock price of the S&P 500 over the last 10 years
MSFT_df.plot(title = "Stock price of the S&P 500 over the last 10 years", figsize = (15,5))
```

## 8. Calculate Daily Returns

```python
# Calculate the daily return using the 'pct_change()' function
MSFT_daily_returns = MSFT_df.pct_change()
```

## 9. Plot Daily Returns

```python
# Plot the daily returns of MSFT over the last 10 years
MSFT_daily_returns.plot(title = "Daily returns of MSFT over the last 10 years", figsize = (15,5))
```

## 10. Calculate Cumulative Returns

```python
# Calculate the cumulative returns using the 'cumprod()' function
MSFT_cumulative_returns = (1 + MSFT_daily_returns).cumprod() - 1
MSFT_cumulative_returns
```

## 11. Plot Cumulative Returns

```python
# Plot the daily returns of the S&P 500 over the last 10 years
MSFT_cumulative_returns.plot(title = "Cumulative Returns of MSFT over the last 10 years", figsize = (15,5))
```
