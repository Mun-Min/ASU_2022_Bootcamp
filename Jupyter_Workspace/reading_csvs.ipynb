```python
import pandas as pd 
from pathlib import Path 
```

```python
csv_path = Path("sales_no_header.csv") 
```

```python
# read the CSV file into a dataframe using pandas 
sales_df = pd.read_csv(csv_path, header = None)  # brings header row into first row 
sales_df.tail()
```

```python
# rewrite column names 
columns_header = ["Full Name", "Email", "Address", "Zipcode", "Credit_Card", "Sales_Price"] 
columns_header
```

```python
sales_df.columns = columns_header
sales_df.head()
```

```python
sales_df.info()
```

```python
sales_df.describe()
```

```python
# rename columns 
sales_df = sales_df.rename(columns = {
    "Full Name" : "full_name", 
    "Credit_Card" : "credit_card_number" 
})

sales_df.head() 
```

```python
# reorder columns 
sales_df = sales_df[['credit_card_number', 'Sales_Price', 'full_name', 'Email', 'Address', 'Zipcode']] 
sales_df.head() 
```

```python
sales_df.columns
```

```python
# deleting a column 
sales_df = sales_df.drop(columns = ["Email"]) 
sales_df.head() 
```
